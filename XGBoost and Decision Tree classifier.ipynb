```python
def convert_to_minutes(time_value):
    if pd.isnull(time_value):
        return None

    if isinstance(time_value, str):
        time_str = time_value.replace('PT', '')

        hours, minutes = 0, 0
        if 'H' in time_str:
            hours = int(time_str.split('H')[0])
            time_str = time_str.split('H')[1]

        if 'M' in time_str:
            minutes = int(time_str.split('M')[0])

        total_minutes = hours * 60 + minutes
        return total_minutes
    else:
        return time_value  # Return the value as it is (no conversion)

recipes['CookTime'] = recipes['CookTime'].apply(convert_to_minutes)
recipes['PrepTime'] = recipes['PrepTime'].apply(convert_to_minutes)
recipes['TotalTime'] = recipes['TotalTime'].apply(convert_to_minutes)
```

```python
from fractions import Fraction
def convert_to_decimal(value):
    try:
        return float(value)
    except ValueError:
        match = re.match(r'(\d+) (\d+)/(\d+)', str(value))
        if match:
            whole, numerator, denominator = map(int, match.groups())
            return whole + numerator / denominator
        else:
            return pd.NaT

recipes_cleaned['RecipeYield'] = recipes_cleaned['RecipeYield'].apply(convert_to_decimal)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from string import punctuation

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Tokenization and Text Cleaning
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def preprocess_text(text):
    if isinstance(text, str):  # Check if the value is a string
        # Tokenization
        tokens = word_tokenize(text.lower())  # Convert to lowercase

        # Remove punctuation and stop words
        tokens = [token for token in tokens if token not in punctuation and token not in stop_words]

        # Stemming
        stemmed_tokens = [stemmer.stem(token) for token in tokens]

        return stemmed_tokens
    else:
        return []  # Return an empty list if the value is not a string or missing


```

```python
# Apply the preprocess_text function to your dataset column containing text
recipes_cleaned['RecipeInstructions'] = recipes_cleaned['RecipeInstructions'].apply(preprocess_text)
recipes_cleaned['Cuisine'] = recipes_cleaned['Cuisine'].apply(preprocess_text)
recipes_cleaned['Keywords'] = recipes_cleaned['Keywords'].apply(preprocess_text)
recipes_cleaned['RecipeCategory'] = recipes_cleaned['RecipeCategory'].apply(preprocess_text)
```

```python
# Concatenate the preprocessed text data into strings
recipes_cleaned['RecipeInstructions'] = recipes_cleaned['RecipeInstructions'].apply(lambda x: ' '.join(x))
recipes_cleaned['Cuisine'] = recipes_cleaned['Cuisine'].apply(lambda x: ' '.join(x))
recipes_cleaned['Keywords']=recipes_cleaned['Keywords'].apply(lambda x: ' '.join(x))
recipes_cleaned['RecipeCategory'] = recipes_cleaned['RecipeCategory'].apply(lambda x: ' '.join(x))
```

```python
from scipy.sparse import hstack
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack

# Combine numerical columns with text data
combined_data = recipes_cleaned[['Calories', 'Duration', 'IsHealthy', 'FreeOf', 
                                 'HighIn', 'LowIn', 'Num_Images', 'DescriptionWordCount', 'ReviewCount',
                                'FatContent', 'SaturatedFatContent', 'CholesterolContent','SodiumContent', 'CarbohydrateContent',
                                'FiberContent', 'SugarContent', 'ProteinContent', 'RecipeServings','RecipeYield']].copy()
combined_data['text'] = recipes_cleaned['RecipeInstructions'] + ' ' + recipes_cleaned['Cuisine'] + ' ' + \
                        recipes_cleaned['RecipeCategory'] + ' '+recipes_cleaned['Keywords']


from sklearn.impute import SimpleImputer
from scipy.sparse import csr_matrix, hstack
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder
# Handle missing values in numerical features separately
numerical_columns = ['Calories', 'Duration', 'IsHealthy', 'FreeOf', 
                     'HighIn', 'LowIn', 'Num_Images', 'DescriptionWordCount', 'ReviewCount',
                     'FatContent', 'SaturatedFatContent', 'CholesterolContent', 'SodiumContent','CarbohydrateContent',
                     'FiberContent', 'SugarContent', 'ProteinContent', 'RecipeServings', 'RecipeYield']

imputed_numerical_features = []

for col in numerical_columns:
    imputer = SimpleImputer(strategy='mean')
    imputed_column = imputer.fit_transform(combined_data[[col]])
    imputed_numerical_features.append(imputed_column)

# Combine imputed numerical features
numerical_features_imputed = np.hstack(imputed_numerical_features)

# Convert numerical features to sparse matrices
numerical_features_sparse = csr_matrix(numerical_features_imputed.astype(float))



```

```python
from sklearn.model_selection import train_test_split
# Split the data into features (X) and target (y)
X_text = combined_data['text']
y = recipes_cleaned['AggregatedRating']

# Split the data into train and test sets for text data
X_text_train, X_text_test, y_train, y_test = train_test_split(X_text, y, test_size=0.2, random_state=42)

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit TF-IDF Vectorizer on training text data and transform both training and testing text data
X_tfidf_text_train = tfidf_vectorizer.fit_transform(X_text_train)
X_tfidf_text_test = tfidf_vectorizer.transform(X_text_test)

# Combine TF-IDF with numerical features for both training and testing data
X_train = hstack([X_tfidf_text_train, numerical_features_sparse[:X_tfidf_text_train.shape[0], :]])
X_test = hstack([X_tfidf_text_test, numerical_features_sparse[X_tfidf_text_train.shape[0]:, :]])
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, average_precision_score, roc_auc_score, confusion_matrix, classification_report
model_xgb = XGBClassifier(n_estimators=100, max_depth=10, learning_rate=0.1)

model_xgb.fit(X_train, y_train)

y_pred_xgb = model_xgb.predict(X_test)

accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
conf_matrix_xgb = confusion_matrix(y_test, y_pred_xgb)
class_report_xgb = classification_report(y_test, y_pred_xgb)

print("Accuracy (XGBoost):", accuracy_xgb)
print("Confusion Matrix (XGBoost):\n", conf_matrix_xgb)
print("Classification Report (XGBoost):\n", class_report_xgb)
```

```python
y_pred_proba_xgb = model_xgb.predict_proba(X_test)[:, 1] 

# Calculate evaluation metrics
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
avg_precision_xgb = average_precision_score(y_test, y_pred_proba_xgb)
auc_xgb = roc_auc_score(y_test, y_pred_proba_xgb)

print("Accuracy (XGBoost):", accuracy_xgb)
print("Average Precision (XGBoost):", avg_precision_xgb)
print("AUC (XGBoost):", auc_xgb)
```
